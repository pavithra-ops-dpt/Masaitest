```python
import pandas as pd

# Load dataset
df = pd.read_csv("used_cars_messy.csv")

# Basic exploration
print(df.shape)
print(df.info())
print(df.describe())
```

```python
df = df.dropna(subset=['selling_price'])


fill_values = {}

if 'mileage' in df.columns:
    fill_values['mileage'] = df['mileage'].mode()[0]

if 'engine' in df.columns:
    fill_values['engine'] = df['engine'].mode()[0]

if 'power' in df.columns:
    fill_values['power'] = df['power'].mode()[0]

df.fillna(fill_values, inplace=True)


df['brand'] = df['brand'].str.strip().str.lower()


df['mileage'] = df['mileage'].str.extract(r'(\d+\.?\d*)')
df['mileage'] = df['mileage'].astype(float)


df = df.drop_duplicates()


print(df.info())
```

```python
from sklearn.metrics import mean_absolute_error

mean_price = df['selling_price'].mean()

y_true = df['selling_price']
y_pred = [mean_price] * len(df)

mae = mean_absolute_error(y_true, y_pred)

print("Baseline Mean Price:", mean_price)
print("Baseline MAE:", mae)
```
